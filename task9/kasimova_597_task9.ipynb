** Задача 3 (5%) **
Докажите, что the k-means всегда сходится.


Алгоритм k-means состоит из двух шагов: (инициализируем центры кластеров $\mu_y$)
* относим каждый объект к кластеру $y_i := argmin_{y_i \in Y}\rho(x_i, \mu_y)$
* вычисление новых положений центров

Эти шаги мы повторяем, пока метки кластеров для каждого объекта не перестанут изменяться

Почему алгоритм остановится: на каждом шаге мы делаем разбиение конечного множества, а их (разбиений) конечное число. Алгоритм не зациклится, потому что как известно среднее минимизирует MSE, а каждый центр мы выбираем как среднее взвешенное расстояние, то есть на каждом шаге мы поддерживает инвариант того, что среднее квадратичное отклонение уменьшается, но оно также ограниченно снизу нулём, значит, центры кластеров в какой-то момент перестанут меняться.

```python
from imageio import imread
from scipy import ndimage
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, MeanShift, AgglomerativeClustering, estimate_bandwidth
from skimage.io import imread
from sklearn.neighbors import kneighbors_graph
from skimage.measure import compare_ssim
```

```python
url = 'two-lorikeet-birds-2293918.jpg'
image = imread(url)
plt.figure(figsize=(15,10))
plt.imshow(image)
image.shape
```

```python
X = image.reshape(-1,3)
X.shape
```

```python
X = image.reshape(275*400, 3)
X.shape
```

```python
# your code here
def create_color_dict(image, n):
    colors_for_clusters = {}
    R = np.array(X)[:,0]
    G = np.array(X)[:,1]
    B = np.array(X)[:,2]
    for cluster_number in range(n):
        class_member_mask = (image == cluster_number)
        colors_for_clusters[cluster_number] = (
            R[class_member_mask].mean(),
            G[class_member_mask].mean(),
            B[class_member_mask].mean())
    return colors_for_clusters
```

```python
def create_new_image(image, colors_for_clusters):
    res = []
    for i in image:
        if i == -1:
            res.append((0, 0, 0))
        else:
            res.append(tuple(colors_for_clusters[i]))
    res = np.array(res).reshape(275, 400, 3)
    return res
```

```python
def draw(res, k):
    colors = create_color_dict(res, k)
    new_image = create_new_image(res, colors)
    plt.imshow(new_image[:,:,0])
    plt.show()
```

```python
def apply_kmeans(k):
    km = KMeans(n_clusters=k)
    res = km.fit_predict(X)
    draw(res, k)
    return create_new_image(res, create_color_dict(res, k))
```

```python
km2 = apply_kmeans(2)
km3 = apply_kmeans(3)
km10 = apply_kmeans(10)
km20 = apply_kmeans(20)
```

```python
from sklearn.datasets.samples_generator import make_blobs
X, _ = make_blobs(n_samples=len(X), centers=X, shuffle=False)
```

```python
from collections import Counter
```

```python
bandwidth = estimate_bandwidth(X, quantile=0.05, n_samples=500)
ms = MeanShift(bandwidth=bandwidth, bin_seeding=True)
res_ms = ms.fit_predict(X)
draw(res_ms, len(Counter(res_ms)))
```

```python
res_ms = create_new_image(res_ms, create_color_dict(res_ms, len(Counter(res_ms))))
```

```python
knn_graph = kneighbors_graph(X, 30, include_self=False)

def apply_agl(k):
    km = AgglomerativeClustering(n_clusters=k, connectivity=knn_graph)
    res = km.fit_predict(X)
    draw(res, k)
    return create_new_image(res, create_color_dict(res, k))
```

```python
agl2 = apply_agl(2)
agl3 = apply_agl(3)
agl10 = apply_agl(10)
agl20 = apply_agl(20)
```
