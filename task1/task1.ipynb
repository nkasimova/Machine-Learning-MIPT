–ù–∞—á–Ω–µ–º —Å —Ç–æ–≥–æ, —á—Ç–æ –ø–æ—Å–º–æ—Ç—Ä–∏–º –Ω–∞ –ø—Ä–æ—Ü–µ—Å—Å –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è ROC-AUC –∫—Ä–∏–≤–æ–π. –î–ª—è—è —ç—Ç–æ–≥–æ –º—ã –±–µ—Ä–µ–º –∫–≤–∞–¥—Ä–∞—Ç —Å–æ —Å—Ç–æ—Ä–æ–Ω–∞–º–∏ –µ–¥–∏–Ω–∏—á–Ω–æ–π –¥–ª–∏–Ω—ã, —Ä–∞–∑–±–∏–≤–∞–µ–º –µ–≥–æ –Ω–∞ m(—á–∏—Å–ª–æ –µ–¥–∏–Ω–∏—Ü —Å—Ä–µ–¥–∏ –º–µ—Ç–æ–∫) —Ä–∞–≤–Ω—ã—Ö —á–∞—Å—Ç–µ–π –≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—å–Ω—ã–º–∏ –ª–∏–Ω–∏—è–º–∏ –∏ –Ω–∞ n(—á–∏—Å–ª–æ –Ω—É–ª–µ–π —Å—Ä–µ–¥–∏ –º–µ—Ç–æ–∫) —Ä–∞–≤–Ω—ã—Ö —á–∞—Å—Ç–µ–π –≤–µ—Ä—Ç–∏–∫–∞–ª—å–Ω—ã–º–∏ –ª–∏–Ω–∏—è–º–∏. –ü–æ–ª—Ü—á–∞–µ—Ç—Å—è —Å–µ—Ç–∫–∞ –∏–∑ –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫–æ–≤. –ò–∑–Ω–∞—á–∞–ª—å–Ω–æ –º—ã –Ω–∞–∑–æ–¥–∏–º—Å—è –≤ —Ç–æ—á–∫–µ (0, 0). –î–∞–ª–µ–µ, –¥–µ–ª–∞–µ–º 1 —à–∞–≥ –≤–≤–µ—Ä—Ö, –µ—Å–ª–∏ –º–µ—Ç–∫–∞ —Ä–∞–≤–Ω–∞ 1, –∏ —à–∞–≥ –≤–ø—Ä–∞–≤–æ, –µ—Å–ª–∏ –º–µ—Ç–∫–∞ —Ä–∞–≤–Ω–∞ 0. –¢.–µ. –ø–æ–ª—É—á–∞–µ—Ç—Å—è, —á—Ç–æ –¥–ª—è —Å–ª—É—á–∞–π–Ω–æ–≥–æ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ç–æ—Ä–∞ –º—ã –º–æ–∂–µ–º –æ–∂–∏–¥–∞—Ç—å –ø—Ä–∏–º–µ—Ä–Ω–æ –æ–¥–∏–Ω–∞–∫–æ–≤–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω—É–ª–µ–π –∏ –µ–¥–∏–Ω–∏—Ü. –ü–æ—ç—Ç–æ–º—É –∫—Ä–∏–≤–∞—è –±—É–¥–µ—Ç –∏–º–µ—Ç—å –≤–∏–¥ –ø—Ä—è–º–æ–π y=x. –ò –ø–ª–æ—â–∞–¥—å –ø–æ–¥ –Ω–µ–π, —Å–æ–æ—Ç–≤–µ—Ç—Å–≤–µ–Ω–Ω–æ, —Ä–∞–≤–Ω–∞ 0.5.

** –ó–∞–¥–∞—á–∞ 2**
–ü–æ–∫–∞–∂–∏—Ç–µ, —á—Ç–æ —Å —Ä–æ—Å—Ç–æ–º —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç–∏ –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–∞ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –ø—Ä–∏ —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–∏ —Ç–æ—á–µ–∫ –≤ –∫—É–±–µ $[0; 1]^ùëë$ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ø–∞—Å—Ç—å –≤ –∫—É–± $[0; 0, 99]^ùëë$ —Å—Ç—Ä–µ–º–∏—Ç—Å—è –∫ –Ω—É–ª—é. –≠—Ç–æ –æ–¥–Ω–∞ –∏–∑ –∏–ª–ª—é—Å—Ç—Ä–∞—Ü–∏–π –ø—Ä–æ–∫–ª—è—Ç–∏—è —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç–µ–π (dimension curse). –ü–æ–ø—Ä–æ–±—É–π—Ç–µ –ø—Ä–∏–¥—É–º–∞—Ç—å –∏–ª–∏ –Ω–∞–π—Ç–∏ –µ—â–µ –∫–∞–∫—É—é-–Ω–∏–±—É–¥—å –∏–ª–ª—é—Å—Ç—Ä–∞—Ü–∏—é –∫ —ç—Ç–æ–º—É —è–≤–ª–µ–Ω–∏—é –∏ –∫—Ä–∞—Ç–∫–æ –∏–∑–ª–æ–∂–∏—Ç—å.

**–†–µ—à–µ–Ω–∏–µ:**
–ö—É–± —Å –¥–ª–∏–Ω–æ–π —Ä–µ–±—Ä–∞ 0.99 –±—É–¥–µ—Ç —Å–æ—Å—Ç–∞–≤–ª—è—Ç—å $0.99^d$ —á–∞—Å—Ç—å –æ—Ç –æ–±—ä–µ–º–∞ –µ–¥–∏–Ω–∏—á–Ω–æ–≥–æ –∫—É–±–∞. 
–¢.–µ. –ø—Ä–∏ $d \to \infty$ –æ–±—ä–µ–º –º–∞–ª–µ–Ω—å–∫–æ–≥–æ –∫—É–±–∞ —Å—Ç—Ä–µ–º–∏—Ç—Å—è –∫ –Ω—É–ª—é: $\lim_{d \to \infty}{0.99^d}=0$. 
–ò–ª–∏, –±–æ–ª–µ–µ —Ñ–æ—Ä–º–∞–ª—å–Ω–æ, –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã –∫–∞–∂–¥–æ–π —Ç–æ—á–∫–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω—ã –Ω–µ–∑–∞–≤–∏—Å–∏–º–æ. –ü–æ—ç—Ç–æ–º—É –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ø–∞—Å—Ç—å –≤ –∫—É–± $[0; 0.99]^ùëë$ - —ç—Ç–æ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ø–∞—Å—Ç—å –≤ –æ—Ç—Ä–µ–∑–æ–∫ $[0; 0.99]$. –¢.–µ. –∏–º–µ–µ—Ç –º–µ—Å—Ç–æ —Ä–∞–≤–µ–Ω—Å—Ç–≤–æ:
$$
P([0;0.99]^d) = \prod_{n=1}^d P([0; 0.99])=0.99^d \to_{d \to \infty} 0
$$

–í –∫–∞—á–µ—Å—Ç–≤–µ –Ω–µ–º–Ω–æ–≥–æ –¥—Ä—É–≥–æ–≥–æ –ø—Ä–∏–º–µ—Ä–∞ —Ä–∞—Å—Å–º–æ—Ç—Ä–∏–º —à–∞—Ä—ã —Ä–∞–¥–∏—É—Å–æ–≤ 1 –∏ 0.9 –≤ n-–º–µ—Ä–Ω–æ–º –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–µ.
–° —É—á–µ—Ç–æ–º —Ç–æ–≥–æ, —á—Ç–æ –≤ n-–º–µ—Ä–Ω–æ–º –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–µ —Ñ–æ—Ä–º—É–ª–∞ –æ–±—ä–µ–º–∞ —à–∞—Ä–∞ —Ä–∞–¥–∏—É—Å–∞ R –∏–º–µ–µ—Ç –≤–∏–¥
$$
V_n(R) = \frac{\pi^{\frac{n}{2}}}{–ì(\frac{n}{2} + 1)}R^n
$$
–≥–¥–µ –ì() - —ç–π–ª–µ—Ä–æ–≤—Å–∫–∞—è –≥–∞–º–º–∞-—Ñ—É–Ω–∫—Ü–∏—è.

–¢–æ –æ—Ç–Ω–æ—à–µ–Ω–∏–µ –æ–±—ä–µ–º–æ–≤ —ç—Ç–∏—Ö —à–∞—Ä–æ–≤ –±—É–¥–µ—Ç —Ä–∞–≤–Ω–æ n-–æ–π —Å—Ç–µ–ø–µ–Ω–∏ –æ—Ç–Ω–æ—à–µ–Ω–∏—è –∏—Ö —Ä–∞–¥–∏—É—Å–æ–≤, —Ç.–µ.
$$
\frac{V_n(0.9)}{V_n(1)} = \left( \frac{0.9}{1}\right)^n = 0.9^n \rightarrow_{n \to \infty} 0
$$
–¢.–µ. –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ø–∞—Å—Ç—å –≤ —à–∞—Ä —Ä–∞–¥–∏—É—Å–∞ $0.9^n$ –±—É–¥–µ—Ç —Å—Ç—Ä–µ–º–∏—Ç—å—Å—è –∫ –Ω—É–ª—é, –ø—Ä–∏ $n \to \infty$ 

<h1 align="center">–ò—Å—Å–ª–µ–¥—É–π—Ç–µ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å Bias-Variance –¥–ª—è kNN (3 –±–∞–ª–ª–∞) </h1> 

–ò—Å–ø–æ–ª—å–∑—É—è –¥–∞–Ω–Ω—ã–µ –∏–∑ –∑–∞–¥–∞—á–∏ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ –¥–æ—Ö–æ–¥–æ–≤ –∏–Ω–¥–∏–≤–∏–¥—É—É–º–∞ http://archive.ics.uci.edu/ml/machine-learning-databases/adult, –ø–æ—Å—Ç—Ä–æ–π—Ç–µ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –≤–µ–ª–∏—á–∏–Ω bias, variance –∏ noise –¥–ª—è —Å–ª–µ–¥—É—é—â–∏—Ö –ø–∞—Ä "–º–æ–¥–µ–ª—å|–ø–∞—Ä–∞–º–µ—Ç—Ä".
–ü—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∞ –¥–∞–Ω–Ω—ã—Ö (–ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤, —Ä–∞–∑–º–µ—Ç–∫–∞ –∫–ª–∞—Å—Å–æ–≤) –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç—Å—è –≤–∞–º–∏. 

* kNN, $n\_neigbours$ - —á–∏—Å–ª–æ —Å–æ—Å–µ–¥–µ–π –ø—Ä–∏ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏;
* $p$ - —Å—Ç–µ–ø–µ–Ω—å –º–µ—Ç—Ä–∏–∫–∏ –ú–∏–Ω–∫–æ–≤—Å–∫–æ–≥–æ;
* $\ell\_train$ - –¥–ª–∏–Ω–∞ –æ–±—É—á–∞—é—â–∏–π –≤—ã–±–æ—Ä–∫–∏;

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn import metrics, neighbors, dummy, preprocessing, svm
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn import linear_model
from sklearn.neighbors import KNeighborsClassifier
from tqdm import tqdm as tqdm
import pandas as pd
%matplotlib inline
```

```python
# Settings
n_repeat = 100       # Number of iterations for computing expectations
test_size = 0.5       # (Relative) Size of the test set

# –ò—Å—Å–ª–µ–¥—É–µ–º–∞—è —Å–µ—Ç–∫–∞ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤
# –ø–æ—è—Å–Ω–µ–Ω–∏–µ: –¥–ª–∏–Ω–∞ train –≤—ã–±–æ—Ä–∫–∏ –¥–æ–ª–∂–Ω–∞ –≤–∞—Ä—å–∏—Ä–æ–≤–∞—Ç—å—Å—è –æ—Ç 50% –¥–æ 200% –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ –¥–ª–∏–Ω—ã —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∏  
parameters = {"k": range(20), "p":  np.arange(1, 10, 0.5), "l_train": np.arange(0.5, 2, 0.1)}

# Get Data
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data'

df = pd.read_csv(url, header=None)
columns = ('age workclass fnlwgt education educ-num marital-status occupation relationship '
           'race sex capital-gain capital-loss  hours-per-week native-country salary')
df.columns = columns.split()

df['salary'] = df['salary'].apply((lambda x: x==' >50K'))
```

```python
df.head()
```

```python
#—Å—Ç–æ–ª–±—Ü—ã –¥–ª—è –º–∞—Ç—Ä–∏—Ü—ã –ø—Ä–∏–∑–Ω–∞–∫–æ–≤, –æ—Å—Ç–∞–≤–∏–ª–∏ —Ç–æ–ª—å–∫–æ —Å —Ü–µ–ª–æ—á–∏—Å–ª–µ–Ω–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏
charact_matrix = ['age', 'fnlwgt', 'educ-num', 'capital-gain', 'capital-loss', 'hours-per-week']
```

```python
def get_split_data(l_train):
    # —Ä–∞–∑–±–µ–π—Ç–µ –¥–∞–Ω–Ω—ã–µ –Ω–∞ —Ç–µ—Å—Ç –∏ —Ç—Ä–µ–π–Ω
    train, test = train_test_split(df, test_size=1/(1 + l_train))
    X_train = []
    y_train = []
    # –î–ª—è —Ç–æ–≥–æ —á—Ç–æ–±—ã –æ—Ü–µ–Ω–∏—Ç—å bias –∏ variance, –Ω—É–∂–Ω–æ —Å–≥–µ–Ω–µ—Ä–∏—Ä–æ–≤–∞—Ç—å 
    # —Å–ª—É—á–∞–π–Ω—ã–µ –ø–æ–¥–≤—ã–±–æ—Ä–∫–∏ —Å –≤–æ–∑–≤—Ä–∞—â–µ–Ω–∏–µ–º –∏–∑ –∏—Å—Ö–æ–¥–Ω–æ–≥–æ –º–∞—Å—Å–∏–≤–∞ –¥–∞–Ω–Ω—ã—Ö 
    # (–±—É—Ç—Å—Ç—Ä–µ–ø–∏–Ω–≥)
    for i in range(n_repeat):
        sample_train = train.sample(n=len(train), replace=True) 
        #X, y = —Ä–∞–∑–±–µ–π—Ç–µ sample_train –Ω–∞ –º–∞—Ç—Ä–∏—Ü—É –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –∏ —Ü–µ–ª–µ–≤–æ–π –≤–µ–∫—Ç–æ—Ä
        X = np.array(sample_train[charact_matrix])
        y = np.array(sample_train['salary'])
        X_train.append(X)
        y_train.append(y)
    X_test = np.array(test[charact_matrix])
    y_test = np.array(test['salary'])

    return X_train, y_train, X_test, y_test
```

```python
def show_plot(bias_variance_df):
    # –Ω–∞—Ä–∏—Å—É–π—Ç–µ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å bias, variance –∏ error –æ—Ç –∑–Ω–∞—á–µ–Ω–∏–π –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤
    plt.figure(figsize=(10, 8))
    bias_variance_df.set_index(p_name).plot()
    plt.show()
```

```python
# –î–ª—è –∫–∞–∂–¥–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è –ø–∞—Ä–∞–º–µ—Ä–∞ –æ—Ü–µ–Ω–∏—Ç–µ –∑–Ω–∞—á–µ–Ω–∏–µ error, bias –∏ variance —Å–æ–≥–ª–∞—Å–Ω–æ —Å–ª–µ–¥—É—é—â–µ–π —Å—Ö–µ–º–µ
for p_name in parameters.keys():
    X_train, y_train, X_test, y_test = get_split_data(test_size)
    bias_variance_df = pd.DataFrame.from_dict({p_name:parameters[p_name],'bias':list([np.NaN]*len(parameters[p_name]))
                                               , 'variance':list([np.NaN]*len(parameters[p_name]))
                                               , 'error':list([np.NaN]*len(parameters[p_name]))})
    #–ü—Ä–æ–∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º n_test
    n_test = len(y_test)
    for parameter in tqdm(parameters[p_name]):
        if (p_name == 'l_train'):
            X_train, y_train, X_test, y_test = get_split_data(parameter)
            n_test = len(y_test)
        
        # Compute predictions
        y_predict = np.zeros((n_test, n_repeat))
        
        for i in range(n_repeat):
            #–î–æ–±–∞–≤–∏–ª–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä
            if (p_name == 'k'):
                estimator = KNeighborsClassifier(n_neighbors=parameter+1).fit(X_train[i], y_train[i])
            elif (p_name == 'p'):
                estimator = KNeighborsClassifier(p=parameter, n_neighbors=2).fit(X_train[i], y_train[i])
            else:
                estimator = KNeighborsClassifier().fit(X_train[i], y_train[i])
                
            y_predict[:, i] = estimator.predict(X_test)

        # Bias^2 + Variance + Noise decomposition of the mean squared error
        y_error = np.zeros(n_test)

        for i in range(n_repeat):
            y_error += (y_test - y_predict[:, i]) ** 2

        y_error /= (n_repeat)

        # –í –∫–∞–∂–¥–æ–π —Ç–æ—á–∫–µ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∏ bias –ø–æ–ª–æ–∂–∏–º —Ä–∞–≤–Ω—ã–º –∫–≤–∞–¥—Ä–∞—Ç—É —Ä–∞–∑–Ω–æ—Å—Ç–∏ —Ä–µ–∞–ª—å–Ω–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è 
        # –∏ —Å—Ä–µ–¥–Ω–µ–≥–æ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è –ø–æ –≤—Å–µ–º –æ–±—É—á–∞—é—â–∏–º –≤—ã–±–æ—Ä–∫–∞–º.
        y_bias = (y_test - np.mean(y_predict, axis=1)) ** 2

        # Variance –ø–æ–ª–æ–∂–∏–º —Ä–∞–≤–Ω—ã–º —Å—Ä–µ–¥–Ω–µ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–º—É —Ä–∞–∑–±—Ä–æ—Å—É –∑–Ω–∞—á–µ–Ω–∏–π –ø–æ –≤—Å–µ–º –æ–±—É—á–∞—é—â–∏–º –≤—ã–±–æ—Ä–∫–∞–º. 
        # –ò—Ç–æ–≥–æ–≤–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ bias –∏ variance –≤–æ–∑—å–º—ë–º —Ä–∞–≤–Ω—ã–º —Å—Ä–µ–¥–Ω–µ–º—É –ø–æ –≤—Å–µ–º —Ç–æ—á–∫–∞–º —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∏
        y_var = np.var(y_predict, axis=1)

        #—Å–æ—Ö—Ä–∞–Ω–∏—Ç–µ —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ bias^2 –¥–ª—è –¥–∞–Ω–Ω–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è –ø–∞—Ä–∞–º–µ—Ç—Ä–∞
        bias_variance_df.loc[bias_variance_df[p_name]==parameter, 'bias'] = y_bias.mean()

        #—Å–æ—Ö—Ä–∞–Ω–∏—Ç–µ —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ variance –¥–ª—è –¥–∞–Ω–Ω–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è –ø–∞—Ä–∞–º–µ—Ç—Ä–∞
        bias_variance_df.loc[bias_variance_df[p_name]==parameter, 'variance'] = y_var.mean()

        #—Å–æ—Ö—Ä–∞–Ω–∏—Ç–µ —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏ error –¥–ª—è –¥–∞–Ω–Ω–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è –ø–∞—Ä–∞–º–µ—Ç—Ä–∞
        bias_variance_df.loc[bias_variance_df[p_name]==parameter, 'error'] = y_error.mean()
    show_plot(bias_variance_df)
```

Bias, error –∏ variance –≤ –ø–æ–ª—É—á–µ–Ω–Ω—ã—Ö –≤—ã—à–µ –≥—Ä–∞—Ñ–∏–∫–∞—Ö —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—Ç —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏–º –æ–∂–∏–¥–∞–Ω–∏—è–º.  –ò bias, –∏ variance, –∏ error –±–æ–ª—å—à–µ –≤—Å–µ–≥–æ –º–µ–Ω—è—é—Ç—Å—è –ø—Ä–∏ –∏–∑–º–µ–Ω–µ–Ω–∏–∏ —á–∏—Å–ª–∞ —Å–æ—Å–µ–¥–µ–π (k).

<h2 align="center">–†–µ–∞–ª–∏–∑—É–π—Ç–µ kNN (2 –±–∞–ª–ª–∞)</h2>

```python
import warnings
import numpy as np
from scipy.spatial.distance import cosine, euclidean, minkowski

warnings.simplefilter("ignore")
```

```python
class kNNClassifier():
    def __init__(self, n_estimators, metric):
        """
       Parameters
        ----------
        n_neighbours: int
            –ß–∏—Å–ª–æ —Å–æ—Å–µ–¥–µ–π#

        metric: *alias
            –º–µ—Ç—Ä–∏–∫–∞ –∏–∑–º–µ—Ä–µ–Ω–∏—è —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–π
        """
        self.n_neighbours = n_estimators
        self.metric = metric

    def fit(self, X, y):
        """
        Parameters
        ----------
        X: 2d np.array
        y: 1d np.array
        """

        # –¢—É—Ç —Ö—Ä–∞–Ω–∏—Ç–µ –æ–ø–∏—Å–∞–Ω–∏–µ –æ–±—ä–µ–∫—Ç–æ–≤ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–∏
        self.X_learn = X

        # –¢—É—Ç —Ö—Ä–∞–Ω–∏—Ç–µ –∑–¥–µ—Å—å –æ—Ç–≤–µ—Ç—ã –ø–æ –∫–∞–∂–¥–æ–º—É –æ–±—ä–µ–∫—Ç—É –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–∏
        self.y_learn = y
        
        # –¢–∞–∫–∂–µ –±—É–¥–µ–º —Ö—Ä–∞–Ω–∏—Ç—å —á–∏—Å–ª–æ –∫–ª–∞—Å—Å–æ–≤
        self.classes_number = len(set(y))

        return self

    def predict(self, X):
        """
        Parameters
        ----------
        X: 2d np.array –º–∞—Ç—Ä–∏—Ü–∞ –æ–±—ä–µ–∫—Ç—ã –ø—Ä–∏–∑–Ω–∞–∫–∏ –Ω–∞ –∫–æ—Ç–æ—Ä—ã—Ö –Ω—É–∂–Ω–æ —Å–∫–∞–∑–∞—Ç—å –æ—Ç–≤–µ—Ç

        Returns
        -------
        y_pred: 1d np.array, –í–µ–∫—Ç–æ—Ä –∫–ª–∞—Å—Å–æ–≤ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –æ–±—ä–µ–∫—Ç–∞
        """
        dist = [] # –•—Ä–∞–Ω–∏—Ç–µ —Ç—É—Ç —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è –¥–æ –∫–∞–∂–¥–æ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–∏ 

        #–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –±–ª–∏–∂–∞–π—à–∏—Ö —Å–æ—Å–µ–¥–µ–π –∏–∑ –∫–∞–∂–¥–æ–≥–æ –∫–ª–∞—Å—Å–∞ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞ X 
        nn_numbers = np.zeros((X.shape[0], self.classes_number)) 

        for i in range(X.shape[0]):
            dist.append([])
            for j in range(self.X_learn.shape[0]):
                # =======================================
                # —Ä–∞—Å—Å—á–∏—Ç–∞–π—Ç–µ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –¥–æ –∫–∞–∂–¥–æ–≥–æ –æ–±—ä–µ–∫—Ç–∞ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–∏
                # ======================================

                # –¥–æ–±–∞–≤–ª—è–µ–º –ø–∞—Ä—É - —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ, –∫–ª–∞—Å—Å
                dist[i].append((self.metric(self.X_learn[j], X[i]), self.y_learn[j]))

            # —Å–æ—Ä—Ç–∏—Ä—É–µ–º –ø–æ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—é
            dist[i].sort()

            # –≤—ã–±–∏—Ä–∞–µ–º –ø–µ—Ä–≤—ã–µ n_neighbours —Å–æ—Å–µ–¥–µ–π
            dist[i] = dist[i][:self.n_neighbours]

            # —Å—á–∏—Ç–∞–µ–º, —Å–∫–æ–ª—å–∫–æ –±–ª–∏–∂–∞–π—à–∏—Ö —Å–æ—Å–µ–¥–µ–π –∏–∑ –∫–∞–∂–¥–æ–≥–æ –∫–ª–∞—Å—Å–∞ –∏–º–µ–µ—Ç –æ–±—ä–µ–∫—Ç
            # dist[obj][train_obj][1] —Å–æ–¥–µ—Ä–∂–∏—Ç –∫–ª–∞—Å—Å
            for j in range(self.n_neighbours):
                nn_numbers[i][dist[i][j][1]] += 1


        # =======================================
        # –ø—Ä–µ–¥—Å–∫–∞–∂–∏—Ç–µ –∫–ª–∞—Å—Å –∫–∞–∂–¥–æ–≥–æ –∏–∑ –æ–±—ä–µ–∫—Ç–æ–≤
        # =======================================
        y_pred = [ np.argmax(nn_numbers[i]) for i in range(X.shape[0])]

        return y_pred
```

–†–µ–∞–ª–∏–∑—É–π—Ç–µ –∞–ª–≥–æ—Ä–∏—Ç–º –æ—Ç–±–æ—Ä–∞ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ ADD_DELL –∏ –ø—Ä–∏–º–µ–Ω–∏—Ç–µ –µ–≥–æ –¥–ª—è kNN –Ω–∞ –¥–∞–Ω–Ω—ã—Ö sklearn.datasets.digits(). 
–î–ª—è —ç—Ç–æ–≥–æ –ø—Ä–µ–¥–ª–∞–≥–∞–µ—Ç—Å—è —Ä–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å —Å–ª–µ–¥—É—é—â–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏

```python
from copy import copy
from copy import deepcopy
```

```python
# –æ–±–Ω–æ–≤–ª–µ–Ω–∏–µ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è –æ—à–∏–±–∫–∏
def update_Q_min_index(estimator, Q_min, Q_min_set, Q_min_index, feature_set, 
                       X_train, Y_train, X_test, Y_test, update_equal=False):
    feature_set_size = len(feature_set)
    estimator.fit(X_train[:, feature_set], Y_train)
    error = 1 - accuracy_score(Y_test, estimator.predict(X_test[:, feature_set]))
    if error < Q_min[feature_set_size]:
        Q_min_set[feature_set_size] = feature_set
        Q_min[feature_set_size] = error
        if ((Q_min[feature_set_size] < Q_min[Q_min_index])
                or (update_equal and Q_min[feature_set_size] == Q_min[Q_min_index])):
            Q_min_index = feature_set_size
    return Q_min_index
```

```python
# –ù—É–∂–Ω–æ —Ä–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ
def add_one(estimator, feature_set, X_train, Y_train, X_test, Y_test, 
            look_forward=10, start_feature_set=[]):
    Q_min = {0: float('+inf')}
    Q_min_set = {0: []}
    Q_min_index = 0
    
    # –∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è —Å—Ç–∞—Ä—Ç–æ–≤–æ–≥–æ –Ω–∞–±–æ—Ä–∞ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤     
    if start_feature_set:
        estimator.fit(X_train[:, start_feature_set], Y_train)
        error = 1 - accuracy_score(Y_test, estimator.predict(X_test[:, start_feature_set]))
        Q_min = {len(start_feature_set) : error}
        Q_min_set = {len(start_feature_set) : start_feature_set}
        Q_min_index = len(start_feature_set)
    
    # –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤     
    for feature_set_size in range(len(start_feature_set) + 1, len(feature_set) + 1):
        Q_min[feature_set_size] = float('+inf')
        unused_features = set(feature_set).difference(set(Q_min_set[feature_set_size - 1]))
        unused_features = list(unused_features)
        shuffle(unused_features)
        for feature in unused_features:
            #–û–±–Ω–æ–≤–∏–º —Å–ø–∏—Å–æ–∫ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
            new_feature_set = copy(Q_min_set[feature_set_size - 1])
            new_feature_set.append(feature)
            #–û–±–Ω–æ–≤–∏–º –∑–Ω–∞—á–µ–Ω–∏–µ –æ—à–∏–±–∫–∏ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏
            Q_min_index = update_Q_min_index(estimator, Q_min, Q_min_set, 
                                             Q_min_index, new_feature_set, 
                                             X_train, Y_train, X_test, Y_test)
            
        
        print ('Q_min: %.4lf, set_size: %d, added: %s' % (Q_min[feature_set_size], 
                                                          feature_set_size, 
                                                          Q_min_set[feature_set_size][-1]))        
    
    # –∫—Ä–∏—Ç–µ—Ä–∏–π –æ—Å—Ç–∞–Ω–æ–≤–∞
        if Q_min_index + look_forward <= feature_set_size:
            break
    return Q_min_set[Q_min_index]
```

```python
def del_one(estimator, feature_set, X_train, Y_train, X_test, Y_test, look_forward=10):

#     –∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è
    estimator.fit(X_train[:, feature_set], Y_train)
    error = 1 - accuracy_score(Y_test, estimator.predict(X_test[:, feature_set]))
    Q_min = {len(feature_set): error}
    Q_min_set = {len(feature_set): deepcopy(feature_set)}
    Q_min_index = len(feature_set)
    
#     –∏—Ç–µ—Ä–∞—Ü–∏–æ–Ω–Ω–æ–µ —É–¥–∞–ª–µ–Ω–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
    for feature_set_size in list(range(len(feature_set) - 1, 0, -1)):
        Q_min[feature_set_size] = float('+inf')
        features = copy(Q_min_set[feature_set_size + 1])
        shuffle(features)
        for feature in features:
            # –æ–±–Ω–æ–≤–∏–º —Å–ø–∏—Å–æ–∫ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
            new_feature_set = copy(Q_min_set[feature_set_size + 1])
            new_feature_set.remove(feature)
            # –æ–±–Ω–æ–≤–∏–º –∑–Ω–∞—á–µ–Ω–∏–µ –æ—à–∏–±–∫–∏ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏
            Q_min_index = update_Q_min_index(estimator, Q_min, Q_min_set, Q_min_index, 
                                             new_feature_set, X_train, Y_train, X_test, Y_test)
            
        print ('Q_min: %.4lf, set_size: %d, deleted: %s' % (Q_min[feature_set_size], feature_set_size,
            set(Q_min_set[feature_set_size + 1]).difference(set(Q_min_set[feature_set_size])),
        ))
        
#         –∫—Ä–∏—Ç–µ—Ä–∏–π –æ—Å—Ç–∞–Ω–æ–≤–∞
        if feature_set_size + look_forward <= Q_min_index:
            break
    return Q_min_set[Q_min_index]
```

```python
# –ù–∞–∫–æ–Ω–µ—Ü, —Ä–µ–∞–ª–∏–∑—É–π—Ç–µ ADD-DELL
def add_del(estimator, feature_set, X_train, Y_train, X_test, Y_test, look_forward=10):
    Q_min = None
    Q_min_current = float('+inf')
    start_feature_set = []
    
#     –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏—Ç–µ—Ä–∞—Ü–∏–π, –ø—Ä–∏–≤–µ–¥—à–∏—Ö –∫ —É–≤–µ–ª–∏—á–µ–Ω–∏—é –æ—à–∏–±–∫–∏
    bad_iteration_count = 2
    
    while (Q_min is None) or (Q_min_current < Q_min or bad_iteration_count):
        Q_min = Q_min_current
        start_feature_set = add_one(estimator, feature_set, X_train, Y_train, 
                                    X_test, Y_test, look_forward, 
                                    start_feature_set=start_feature_set)
        start_feature_set = del_one(estimator, start_feature_set, X_train, 
                                    Y_train, X_test, Y_test, look_forward)
        estimator.fit(X_train[:, start_feature_set], Y_train)
        Q_min_current = 1 - accuracy_score(Y_test, estimator.predict(X_test[:, start_feature_set]))
        
        if Q_min_current  >=  Q_min:
            bad_iteration_count -= 1
        print ('Q_min: %.4lf, set_size: %d' % (Q_min_current, len(start_feature_set)))
    return start_feature_set
```
